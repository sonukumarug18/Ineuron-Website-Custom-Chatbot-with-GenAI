## Import all required Libraries



```python
from langchain.document_loaders import UnstructuredURLLoader
from langchain.text_splitter import CharacterTextSplitter
# from langchain.vectorstores import Pinecone
# # import pinecone
from langchain.embeddings import HuggingFaceEmbeddings
from transformers import AutoTokenizer, AutoModelForCausalLM
import transformers
from langchain import HuggingFacePipeline
from huggingface_hub import notebook_login
import torch

from langchain.vectorstores import Chroma
```

## Extract data from url

```python
URLs = [
    "https://ineuron.ai/"
]
```

```python
loader = UnstructuredURLLoader(urls = URLs)
data = loader.load()
```

```python
text_splitter=CharacterTextSplitter(separator='\n',
                                    chunk_size=1000,
                                    chunk_overlap=200)
```

## Chunkins

```python
text_chunks=text_splitter.split_documents(data)
```

```python
len(text_chunks)
```

```python
text_chunks[0]
```

## Embedding model

```python
embeddings=HuggingFaceEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2')
```

```python
query_result = embeddings.embed_query("Hello world")
len(query_result)
```

```python
query_result
```

## ChromaDB

```python
vectordb = Chroma.from_documents(text_chunks, embedding=embeddings, persist_directory='./db')
vectordb.persist()
```

```python
import os

persist_directory = os.path.join(os.getcwd(), "chroma_db")
```

```python
vectordb = Chroma(persist_directory=persist_directory,
                  embedding_function=embeddings)
```

## Create LLM wrapper

```python
notebook_login()
```

```python
# model="daryl149/llama-2-7b-chat-hf"

tokenizer = AutoTokenizer.from_pretrained("daryl149/llama-2-7b-chat-hf",
                                          use_auth_token=True,)


model = AutoModelForCausalLM.from_pretrained("daryl149/llama-2-7b-chat-hf",
                                             device_map='auto',
                                             torch_dtype=torch.float16,
                                             use_auth_token=True,
                                              load_in_8bit=True,
                                              #load_in_4bit=True
                                             )
```

```python
pipe = transformers.pipeline("text-generation",
                model=model,
                tokenizer= tokenizer,
                torch_dtype=torch.bfloat16,
                device_map="auto",
                max_new_tokens = 512,
                do_sample=True,
                top_k=30,
                num_return_sequences=1,
                eos_token_id=tokenizer.eos_token_id
                )
```

```python
llm=HuggingFacePipeline(pipeline=pipe, model_kwargs={'temperature':0})
```

```python
llm.predict("Please provide a concise summary of the Book Harry Potter")
```

```python
from langchain.chains import RetrievalQA
```

```python
qa = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff", retriever=vectordb.as_retriever())
```

```python
query = "Tell me the course price of Full Stack Data Science with Generative AI provide by ineuron"
# query = "Tell me the course price of Data Science Masters Pro 2024 provide by ineuron"
```

```python
print(qa.run(query))
```

```python

from google.colab import userdata
GEMINI_API=userdata.get('gemini_api')
```

```python
from langchain.llms import GooglePalm
```

```python
llm = GooglePalm(google_api_key=GEMINI_API,
                 temperature=0.1)
```

```python
from langchain.chains import RetrievalQA
```

```python
from langchain.memory import ConversationSummaryMemory
```

```python
from langchain.chains import ConversationalRetrievalChain
```

```python
# qa = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff", retriever=vectorstore.as_retriever())
memory = ConversationSummaryMemory(llm=llm, memory_key = "chat_history", return_messages=True)
qa = ConversationalRetrievalChain.from_llm(llm, retriever=vectordb.as_retriever(search_type="mmr", search_kwargs={"k":8}), memory=memory)
```

```python

query = "Tell me the course price of Full Stack Data Science with Generative AI provide by ineuron?"
```

```python


print(qa.run(query))
```
